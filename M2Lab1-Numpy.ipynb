## Array Statistics

###Task: Create a numpy array X, print the sum, mean, std, var, max, min of it

```python
X=np.arange(100)
print(X.sum())
print(X.mean())
print(X.std())
print(X.max())
print(X.var())
print(X.min())
```

###Task: Create a numpy matrix X as n * m, print the sum, mean, std, var, max, min of the 1st row of it.

```python
r1=X[0]
print(r1.sum())
print(r1.mean())
print(r1.std())
print(r1.max())
print(r1.var())
print(r1.min())
```

###Task: Create a numpy matrix X as n * m, print the sum, mean, std, var, max, min of the 1st column of it.

```python
r1=X[:,0]
print(r1.sum())
print(r1.mean())
print(r1.std())
print(r1.max())
print(r1.var())
print(r1.min())
```

## Array Linear Algebra

###Task: Create two numpy matrix X with 2 * 2, and Y with 2 * 2. Practice Linear Algebra operations, such as +, -, *, .T, @, etc. 

```python
X=np.random.randint(0,10,(2,2))
Y=np.random.randint(11,20,(2,2))
print(X+Y)
print(X-Y)
print(X*Y)
print(X.T)
print(Y.T)
print(X@Y.T)
print(Y@X.T)
```
